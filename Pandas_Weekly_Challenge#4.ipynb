## Reading data 

```python
baby_names = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
```

# Question 2

## Delete the columns 'Unnamed: 0' and 'Id'.

```python
# Delete the columns 'Unnamed: 0' and 'Id'

column_to_delete = baby_names.drop(['Unnamed: 0', 'Id'], axis=1)
```

```python
column_to_delete
```

# Question 3

## Group the dataset by name, assign to a variable called names, and sort the dataset by highest to lowest count.

```python
# Group the dataset by name, assign to a variable called names, and sort the dataset by highest to lowest count.

names = column_to_delete.groupby('Name')['Count'].sum().reset_index().sort_values(by='Count', ascending=False)
```

```python
names.head()
```

# Question 4

## How many different names exist in the dataset?

```python
# How many different names exist in the dataset?
total_diff_names = len(baby_names['Name'].unique())
```

```python
total_diff_names
```

# Question 5

## What is the name with most occurrences?

```python
# What is the name with most occurrences?

most_names_occur = column_to_delete.groupby('Name').count().sort_values(by='Count', ascending=False)
```

```python
most_names_occur
```

# Question 6

## What is the standard deviation of count of names?

```python
# What is the standard deviation of count of names?

std_names = column_to_delete['Count'].std()
```

```python
std_names
```
